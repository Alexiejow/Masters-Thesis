```python
url = 'https://demagog.org.pl/wypowiedzi/'
s=Service('./chromedriver')

driver = webdriver.Chrome(service=s)
driver.get(url)
wait = WebDriverWait(driver, 20)



#if element with xpath present, click it
try:
    #find button that contains text "Zaaduj wicej"
    accept_cookie_button = driver.find_element(By.XPATH, '//*[@id="cn-accept-cookie"]')
    # print(load_more_button)
    accept_cookie_button.click()
    print('clicked cookie button')
except:
    pass

n = 0
while n<1000:
    # find all links
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    a_tags = soup.find_all('a', href=re.compile('wypowiedzi/.+'))
    for a in [tag.get('href') for tag in a_tags]:
        # if there is a dict in data with link equal to a
        if not any(d['link'] == a for d in data):
            print(a, 'not in data')
            res = rq.get(a)
            soup = BeautifulSoup(res.content, 'html.parser')
            #find quote
            quote = soup.find('div', class_='dg-post-quote__statement').text
            try:
                date = soup.find('div', class_='dg-post-quote__source').text.split(",")[-1].strip()
            except:
                date = None
            try:
                source = soup.find('div', class_='dg-post-quote__source').find('a').get('href')
            except:
                source = None
            
            data.append({
                'link' : a,
                'quote': quote,
                'date' : date,
                'source' : source,
            })
    try:
        #find button with specified classes BY CLASS
        load_more_button = driver.find_element(By.CLASS_NAME, 'dg-cta.dg-load-more')
        load_more_button.click()
        print('clicked load more button')
    except:
        print('no button found')
    n += 1

    print('length of data:', len(data))

    if n%100 == 0:
        print(f'clicked {n} times')
        

#find the href value of all a elements that contain wypowiedzi/XXX
# a_tags = soup.find_all('a', href=re.compile('wypowiedzi/.+'))
```

```python
df_raw = pd.read_csv('demagog_quotes.csv')
# drop first row
df_raw = df_raw.iloc[1:]
# reset index
df_raw.reset_index(drop=True, inplace=True)
# remove one or multiple leading \n's
df_raw['quote'] = df_raw['quote'].apply(lambda x: re.sub(r'^\n+', '', x))
# remove \xa0
df_raw['quote'] = df_raw['quote'].apply(lambda x: re.sub(r'\xa0+', ' ', x))
# remove two or more \n and everything that follows
df_raw['quote'] = df_raw['quote'].apply(lambda x: x.split('\n\n')[0])
# remove square brackets and everything within them, if ther is "przyp" included
df_raw['quote'] = df_raw['quote'].apply(lambda x: re.sub(r'\[.*?przyp.*?\]|\(.*?przyp.*?\)', '', x, flags=re.IGNORECASE))
# remove any spaces occuring before commas and dots
df_raw['quote'] = df_raw['quote'].apply(lambda x: re.sub(r'\s+([,.])', r'\1', x))
# define emoji pattern removal function
def remove_emojis(text):
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                               u"\U0001F700-\U0001F77F"  # alchemical symbols
                               u"\U0001F780-\U0001F7FF"  # Geometric Shapes Extended
                               u"\U0001F800-\U0001F8FF"  # Supplemental Arrows-C
                               u"\U0001F900-\U0001F9FF"  # Supplemental Symbols and Pictographs
                               u"\U0001FA00-\U0001FA6F"  # Chess Symbols
                               u"\U0001FA70-\U0001FAFF"  # Symbols and Pictographs Extended-A
                               u"\U00002702-\U000027B0"  # Dingbats
                               u"\U000024C2-\U0001F251"
                               "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)

# Remove emojis from the text
df_raw['quote'] = df_raw['quote'].apply(remove_emojis)
# remove any multiple spaces
df_raw['quote'] = df_raw['quote'].apply(lambda x: re.sub(r'\s+', ' ', x))

df_raw['quote']
```

```python
import speech_recognition as sr

# Function to transcribe audio
def transcribe_audio(audio):
    recognizer = sr.Recognizer()
    try:
        # Transcribe the audio
        text = recognizer.recognize_google(audio, language='pl-PL')
    except sr.UnknownValueError:
        # Handle unrecognized audio
        text = ""
    except sr.RequestError as e:
        # Handle request error
        print("Could not complete the request: {0}".format(e))
    return text

# Load the audio file
audio_file = sr.AudioFile('Audio for transcripts.wav')

# Initialize duration and offset
duration = 60  # Duration of chunks in seconds
offset = 0  # Start offset

# Open the audio file
with audio_file as source:
    # Create an audio reader
    reader = sr.Recognizer()
    # Initialize transcribed text
    transcribed_text = ""
    while True:
        # Read a chunk of the audio file
        audio = reader.record(source, duration=duration, offset=offset)
        # Transcribe the audio chunk
        transcribed_chunk= transcribe_audio(audio) 
        transcribed_text += transcribed_chunk + ". "
        print(transcribed_chunk)
        # Increase the offset by duration
        offset += duration
```
